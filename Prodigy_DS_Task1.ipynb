```python
df["Number of Deaths"]=df["Number of Deaths"].str.replace(',', '').astype(int)
```

```python
import matplotlib.pyplot as plt
```

```python
# Scale down by a factor of 1,000
df['Number of Deaths (in tens)'] = df['Number of Deaths'] / 10

# Check the new values
print(df[['Number of Deaths', 'Number of Deaths (in tens)']])
```

```python
df
```

```python
df.isnull().sum()
```

```python
grouped_by_country_name = df.groupby('Country Name')['Number of Deaths (in tens)'].mean().reset_index()
#grouped_by_country_name=grouped_by_country_name.head(10)
plt.figure(figsize=(10,5))
plt.bar(grouped_by_country_name['Country Name'], grouped_by_country_name['Number of Deaths (in tens)'])
plt.title("Bar chart ")
plt.xlabel("Country Name")
plt.ylabel("Number of Deaths (in tens)")
plt.show()
```

```python
df["Death Rate Per 100,000"]=df["Death Rate Per 100,000"].str.replace(',', '')
df["Death Rate Per 100,000"]=df["Death Rate Per 100,000"].str.replace('.','')
df["Death Rate Per 100,000"]=df["Death Rate Per 100,000"].astype(int)
```

```python
# Scale down by a factor of 1,000
df['Death Rate Per 100,000 (in Hundreds)'] = df['Death Rate Per 100,000'] / 1000

# Check the new values
print(df[['Death Rate Per 100,000', 'Death Rate Per 100,000 (in Hundreds)']])
```

```python
df
```

```python
plt.figure(figsize=(10,5))
plt.hist(df["Death Rate Per 100,000 (in Hundreds)"])
plt.title("Histrogram chart")
plt.xlabel("Death Rate Per 100,000 (in Hundreds)")
plt.ylabel("Frequency")
plt.show()
```
